MI satte reaches 89%.  To simplify the calsses, separate these two fields into three categories: `MI`, `out_of_state` and `foreign`

```python
us_states  = ['AL','AK','AZ','AR','CA','CO','CT','DE','FL','GA','HI','ID','IL',
             'IN','IA','KS','KY','LA','ME','MD','MA','MI','MN','MS','MO','MT',
             'NE','NV','NH','NJ','NM','NY','NC','ND','OH','OK','OR','PA','RI',
             'SC','SD','TN','TX','UT','VT','VA','WA','WV','WI','WY']

df_train['is_in_state'] = df_train.state.apply(lambda x: x if x in us_states else 'foreign')
df_train.is_in_state[(df_train.is_in_state != 'foreign') & (df_train.is_in_state != 'MI')] = 'out_of_state'
```

```python
df_train.groupby('is_in_state').compliance.agg(['count', 'sum', 'mean', 'std']).sort_values('count', ascending=False).T
```

__Investigate the relationship of the `discount_amount` & `judgment_amount` fields and `compliance`__

```python
df_train_all[['discount_amount', 'judgment_amount']].describe().T
# uneven distribution w/ discount_amount
```

```python
df_train.groupby('discount_amount').compliance.agg(['count', 'sum', 'mean', 'std']).sort_values('count', ascending=False).T
```

This shown as data leakage: whoever gets discount most likely will pay for the fine.

Separate the values into two separated class w/ binary values: True as paid, False as non-paid

```python
df_train['is_discount'] = df_train.discount_amount.apply(lambda x:1 if x > 0 else 0)
```

```python
# separate the judgement into 3 classes w/ 25%, 50%, 75%
df_train['judgment_level'] = pd.cut(df_train.judgment_amount, bins=[-1, 140, 305, float("inf")])

df_train.groupby('judgment_level').compliance.agg(['count', 'sum', 'mean', 'std']).sort_index(ascending=False).T
```
